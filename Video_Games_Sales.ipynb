```python
df['User_Score'].unique()
df.loc[df['User_Score']=='tbd'] = np.nan
df['User_Score'] = df['User_Score'].astype(float)
df = df[df['Year_of_Release'].notna()]
df['Year_of_Release'] = df['Year_of_Release'].astype(int)
```

## Quin publisher ven mes? 

```python
df_topPublisher_sales = df.groupby(['Publisher'], as_index = False).sum().sort_values('Global_Sales', ascending=False).head(20)
df_topPublisher_sales.drop(columns=['Year_of_Release', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count'], inplace=True)
df_topPublisher_sales
```

## Quina estudii ven mes i és més ben valorat? 

```python
df_topDeveloper_sales = df.groupby(['Developer'], as_index = False).sum().sort_values('Global_Sales', ascending=False).head(20)
df_topDeveloper_sales.drop(columns=['Year_of_Release', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count'], inplace=True)
df_topDeveloper_sales
```

```python
for idx, r in df_topDeveloper_sales.iterrows():
    dev  = r['Developer']

    data = df[df['Developer'] == dev]
    data.dropna(inplace=True)

    data['Weighted Critic Score'] = (data['Critic_Score']/10) * data['Critic_Count']
    data['Weighted User Score'] = data['User_Score'] * data['User_Count']

    total_critic = data['Weighted Critic Score'].sum() / data['Critic_Count'].sum()
    total_user = data['Weighted User Score'].sum() / data['User_Count'].sum()

    df_topDeveloper_sales.loc[idx,'Avg. Weighted Critic Score'] = total_critic
    df_topDeveloper_sales.loc[idx,'Avg. Weighted User Score'] = total_user

df_topDeveloper_sales.set_index('Developer', inplace=True)
df_topDeveloper_sales.sort_values('Avg. Weighted User Score', ascending=False)
```

## Com de ben valorat són els jocs? 

```python
df_gamesRelease = df.groupby(['Year_of_Release'],as_index=False).sum()
df_gamesRelease_count = df.groupby(['Year_of_Release'],as_index=False).size()
df_gamesRelease['Number of games'] = df_gamesRelease_count['size']

df_gamesRelease = df_gamesRelease.drop(columns=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count'])
```

```python
for idx, r in df_gamesRelease.iterrows():
    publisher  = r['Year_of_Release']

    data = df[df['Year_of_Release'] == publisher]
    data.dropna(inplace=True)

    data['Weighted Critic Score'] = (data['Critic_Score']/10) * data['Critic_Count']
    data['Weighted User Score'] = data['User_Score'] * data['User_Count']

    total_critic = data['Weighted Critic Score'].sum() / data['Critic_Count'].sum()
    total_user = data['Weighted User Score'].sum() / data['User_Count'].sum()

    df_gamesRelease.loc[idx,'Avg. Weighted Critic Score'] = total_critic
    df_gamesRelease.loc[idx,'Avg. Weighted User Score'] = total_user

df_gamesRelease.set_index('Year_of_Release', inplace=True)
df_gamesRelease
```
